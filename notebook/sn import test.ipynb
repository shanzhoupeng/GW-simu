```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as sciconsts
import scipy.fftpack as fft
import scipy.interpolate as sci_interpolate
```

```python
Sn=np.loadtxt('ZERO_DET_high_P.txt')
```

```python
Sn_fy=Sn[:,1]
Sn_fy
```

```python
Sn_fx=Sn[:,0]
Sn_fx
```

```python
plt.plot(Sn_fx,Sn_fy)
plt.yscale('log')
plt.xscale('log')
```

```python
plt.show()
```

```python
fsn=sci_interpolate.splrep(Sn_fx,Sn_fy)
fx=np.linspace(10.0,8000.0,num=100000)
fy=sci_interpolate.splev(fx,fsn)
```

```python
plt.plot(fx,fy)
plt.yscale('log')
plt.xscale('log')
```

```python
plt.show()
```

```python
fsn=sci_interpolate.interp1d(Sn_fx,Sn_fy)
fx=np.linspace(10.0,8000.0,num=100000)
fy=fsn(fx)
```

```python
plt.plot(fx,fy)
plt.yscale('log')
plt.xscale('log')
```

```python
plt.show()
```
