```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as sciconsts
import scipy.fftpack as fft
import scipy.integrate as sci_integrate
```

```python
def s(t):
    return 3*np.cos(2*sciconsts.pi*1*t*t*t)
```
